**<h3>Functions**

```python
## Useful to do runs (it's faster than the next one)

def phi(t0, f0, df0, f1, df1, E_MIN, E_MAX, barycentering_data):
    
    '''Given specific f0 and f1, it returns the phase 
       value in the interval [-1,1] for all photons'''

    t = barycentering_data.loc[(E_MIN < barycentering_data['ENERGY']) & (barycentering_data['ENERGY'] < E_MAX), 'TIME']
    phi = np.modf(f0*(t-t0) + 1/2*f1*(t-t0)**2 - 0.2)[0]
    
    ##Phase error
    dt = 1e-6
    dphi = np.sqrt( ((t-t0)*df0)**2 + ((t-t0)**2/2*df1)**2 + ((f0+f1*(t-t0))*dt)**2 )
    
    return np.array(phi), dphi
```

```python
## !! Copy of the previous function, but returning a positive phase in [0,1] !!
## Useful to do figures

def phi_positive(t0, f0, df0, f1, df1, E_MIN, E_MAX, barycentering_data):
    
    '''Given specific f0 and f1, it returns the phase 
       value in the interval [0,1] for all photons'''
    
    t = barycentering_data.loc[(E_MIN < barycentering_data['ENERGY']) & (barycentering_data['ENERGY'] < E_MAX), 'TIME']
    phi = np.modf(f0*(t-t0) + 1/2*f1*(t-t0)**2 - 0.2)[0]
    
    ##Phase error
    dt = 1e-6
    dphi = np.sqrt( ((t-t0)*df0)**2 + ((t-t0)**2/2*df1)**2 + ((f0+f1*(t-t0))*dt)**2 )
    
    return np.array(list(map(lambda x: x+1 if x<0 else x, phi))), np.array(dphi)
```

```python
def Z2n(phi_frac_positive, n):
    N = len(phi_frac_positive) #photons number
    
    Z2n = 0
    for k in range(1, n):
        Z2n = Z2n + np.sum(np.cos(2*np.pi*k*phi_frac_positive))**2 + np.sum(np.sin(2*np.pi*k*phi_frac_positive))**2
    
    return Z2n * (2/N)
```

```python
def Zfinal(X, Y, n, barycentering_data, df0, df1):
    
    Z = np.ndarray((X.shape[0],X.shape[1]))
    
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            phi_frac_positive, dphi = phi_positive(t0, X[i][j], df0, Y[i][j], df1, 100, 230216, barycentering_data)
            Z[i][j] = Z2n(phi_frac_positive, n)

    return Z
```

```python
def f0f1convergence(df0, df1, f0mid, f1mid, f0_half_interval, f1_half_interval, n, barycentering_data):

    f0test = np.arange(f0mid-f0_half_interval*df0, f0mid+f0_half_interval*df0, df0)
    f1test = np.arange(f1mid-f1_half_interval*df1, f1mid+f1_half_interval*df1, df1)

    X,Y = np.meshgrid(f0test, f1test)
    Z = Zfinal(X, Y, n, barycentering_data, df0, df1)
    
    i,j = np.unravel_index(Z.argmax(), Z.shape)

    M = Z[i,j] # max of Z2n function
    f0 = X[i][j]
    f1 = Y[i][j]
    
    return f0, f1, X, Y, Z
```

**<h2>Pulsar characterization**

F0 = 9.6393564453698364,  F1 = -8.889253909687E-12

```python
# Inertia moment
I = 10**45 * 10**-7 # Kg*m^2

# Period
P = 1/f0
dP = df0 * 1/f0
print('P = %.16f +/- %.16f s' %(P,dP))

# Period first time derivative
Pdot = -f1/f0**2
Pdot_12 = -f1*1e12/f0**2
dPdot = np.sqrt((df0 * 2*f1/f0**3)**2 + (df1 * 1/f0**2)**2)
dPdot_12 = np.sqrt((df0 * 2*f1*1e12/f0**3)**2 + (df1 * 1/f0**2)**2)
print('Pdot = (%.16f +/- %.16f) e-12 s/s' %(Pdot_12,dPdot_12))

Edot = -4 * np.pi**2 * I * Pdot / P**3  #Kg*m^2/s^2 = Joule/s
Edot_erg = Edot * 10**7
dEdot = 4 * np.pi**2 * I * np.sqrt( (dPdot/P**3)**2 + (dP*3*Pdot/P**4 )**2 ) * 10**7
print('Edot = (%.7f +/- %.7f) 10^36  erg/s' %(Edot_erg/10**36,dEdot/10**36))
```

```python
B = 5.861434349981159*1e18 * np.sqrt(P*Pdot) # gauss
dB = (3.2 * 10**19 / 2*np.sqrt(P*Pdot) ) * np.sqrt((Pdot*dP)**2+(P*dPdot)**2)
print('B = (%.9f +/- %.9f) Gauss' %(B,dB))
```

```python
# Characteristic age
tau = P/(2*Pdot) / (60*60*24*365)
dtau = 1/2*np.sqrt((1/Pdot*dP)**2 + (-P/Pdot**2*dPdot)**2 ) / (60*60*24*365)
print('tau = (%.3lf+/- %.3lf) y ' %(tau,dtau))
```

```python
f = open(os.path.join(data_dir, "pulsar_catalog.txt"),"r")
lines = f.readlines()

result = []
for x in lines:
    x = x.strip("\n")
    if (len(x)>0):
        if(x[0]!= '#'):
            x = " ".join(x.split())
            x = x.split(' ')
            result.append(x)
f.close()
```

```python
P0 = list(result[i][3] for i in range(len(result)) if result[i][6]!='*')
P1 = list(result[i][6] for i in range(len(result)) if result[i][6]!='*')
P0 = [float(x) for x in P0]
P1 = [float(x) for x in P1]
```

```python
plt.figure(figsize=(16,10))
plt.rcParams.update({'font.size': 20})
plt.scatter(np.array(P0), np.array(P1), color='black', marker='.', alpha=0.7)
plt.plot(P, Pdot, 'ro', label='PSR J2021+3651')

# Straights B constant
for exp in range(9,16,2):
    plt.plot(np.array(P0), (10**exp/(5.9*10**18))**2/np.array(P0), linestyle='--', color='tab:orange', linewidth=0.6)
    plt.text(9, (10**exp/(5.9*10**18))**2/15, 'B$\simeq 10^{%i}$ G'%exp, fontsize=12, rotation=-10, color='orange' )

plt.plot(np.array(P0), (10**7/(5.9*10**18))**2/np.array(P0), linestyle='--', color='tab:orange', linewidth=0.6)
plt.text(0.01, (10**7/(5.9*10**18))**2*55, 'B$\simeq 10^{7}$ G', fontsize=12, rotation=-10, color='orange')    

# Straights tau constant
for i in range(1, 10, 3):
    tau = 10**i
    plt.plot(np.array(P0), np.array(P0)/(2*(60*60*24*365)*tau), linestyle='--', color='tab:green', linewidth=0.6)
    plt.text(0.002, 0.003/(2*(60*60*24*365)*(tau)), r'$\tau \simeq 10^{%i}$ y'%i, fontsize=12, rotation=10, color='green')
    
tau = 10**10
plt.plot(np.array(P0), np.array(P0)/(2*(60*60*24*365)*tau), linestyle='--', color='tab:green', linewidth=0.6)
plt.text(2, 3/(2*(60*60*24*365)*(tau)), r'$\tau \simeq 10^{%i}$ y'%i, fontsize=12, rotation=10, color='green')
    
tau = 10**13
plt.plot(np.array(P0), np.array(P0)/(2*(60*60*24*365)*tau), linestyle='--', color='tab:green', linewidth=0.6)
plt.text(.1, .15/(2*(60*60*24*365)*(tau)), r'$\tau \simeq 10^{%i}$ y'%i, fontsize=12, rotation=10, color='green')    


# Straights Edot constant
for exp in range(20, 40, 5):
    plt.plot(np.array(P0), 10**exp/(4*np.pi**2*I)*np.array(P0)**3, linestyle='--', color='tab:blue', linewidth=0.6)
    if(exp!=20):
        plt.text(0.01, 10**exp/(4*np.pi**2*I)*0.013**3, '$\dot E$ [erg/s]', fontsize=10, rotation=28, color='tab:blue' )


plt.xlim(min(np.array(P0)), max(np.array(P0)))
plt.xscale('log')
plt.yscale('log')
plt.ylim(1e-22, 1e-7)
plt.xlabel('$P_0$ [s]')
plt.ylabel('$\dotP$ [s/s]')
plt.legend(loc='upper left')
plt.savefig(os.path.join(figures_dir, 'P_Pdot_diagram.png'))
plt.show()
```
